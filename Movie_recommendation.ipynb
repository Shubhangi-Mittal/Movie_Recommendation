```python
import re
def extract_keywords(text):
  if isinstance(text, str):
    keywords = re.findall(r'\b[a-zA-Z]+\b', text.lower())
    return keywords
  else:
    return []
df['keywords'] = df['overview'].apply(extract_keywords)
print(df)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(keywords):
  filtered_keywords = [word for word in keywords if word not in stop_words]
  return filtered_keywords
df['keywords'] = df['keywords'].apply(remove_stopwords)
print(df)
```

```python
import pandas as pd
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
movies_dataset = df[['id', 'title', 'keywords', 'original_language', 'Genres', 'release_year']]
print(movies_dataset)
```

```python
movies_dataset['keywords'] = movies_dataset['keywords'].apply(lambda x: " ".join(x))
movies_dataset.head()
```

```python
movies_dataset['keywords'] = movies_dataset['keywords'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(movies_dataset['keywords']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
movies_dataset[movies_dataset['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = movies_dataset[movies_dataset['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_dataset.iloc[i[0]].title)
```

```python
def recommend(movie, genre=None, release_year=None, language=None):
    index = movies_dataset[movies_dataset['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        recommended_movie = movies_dataset.iloc[i[0]]
        if (genre is None or genre.lower() in recommended_movie['Genres'].lower()) or \
           (release_year is None or recommended_movie['release_year'] == release_year) or \
           (language is None or recommended_movie['original_language'].lower() == language.lower()):
            print(f"Title: {recommended_movie.title}")
            print(f"Genre: {recommended_movie.Genres}")
            print(f"Release Year: {recommended_movie.release_year}")
            print(f"Language: {recommended_movie.original_language}")
            print("---")

# Example usage:
recommend('The Lego Movie', genre='Animation', release_year=2014)
```

```python
import ipywidgets as widgets
from IPython.display import display
movie_input = widgets.Text(
    value='',
    placeholder='Enter movie title',
    description='Movie Title:',
    disabled=False
)
genre_input = widgets.Text(
    value='',
    placeholder='Enter genre (optional)',
    description='Genre:',
    disabled=False
)
year_input = widgets.IntText(
    value=None,
    placeholder='Enter release year (optional)',
    description='Year:',
    disabled=False
)
language_input = widgets.Text(
    value='',
    placeholder='Enter language (optional)',
    description='Language:',
    disabled=False
)
recommend_button = widgets.Button(description="Recommend")
output = widgets.Output()
def on_recommend_button_clicked(b):
    with output:
        output.clear_output()
        movie = movie_input.value
        genre = genre_input.value
        year = year_input.value
        language = language_input.value
        if movie:
            try:
                recommend(movie, genre, year, language)
            except IndexError:
                print("Movie not found in the dataset.")
        else:
            print("Please enter a movie title.")
recommend_button.on_click(on_recommend_button_clicked)
display(movie_input, genre_input, year_input, language_input, recommend_button, output)
```

```python
import gradio as gr

def recommend_movies(movie_title, genre, year, language):
    try:
        recommend(movie_title, genre, year, language)
        recommendations = []
        index = movies_dataset[movies_dataset['title'] == movie_title].index[0]
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            recommended_movie = movies_dataset.iloc[i[0]]
            if (genre is None or genre.lower() in recommended_movie['Genres'].lower()) or \
               (year is None or recommended_movie['release_year'] == year) or \
               (language is None or recommended_movie['original_language'].lower() == language.lower()):
                recommendations.append(f"Title: {recommended_movie.title}\nGenre: {recommended_movie.Genres}\nRelease Year: {recommended_movie.release_year}\nLanguage: {recommended_movie.original_language}\n---")

        return "\n".join(recommendations)

    except IndexError:
        return "Movie not found in the dataset."

iface = gr.Interface(
    fn=recommend_movies,
    inputs=[
        gr.Textbox(lines=1, placeholder="Enter movie title"),
        gr.Textbox(lines=1, placeholder="Enter genre (optional)"),
        gr.Number(label="Enter release year (optional)"),
        gr.Textbox(lines=1, placeholder="Enter language (optional)")
    ],
    outputs="text",
    title="Movie Recommendation App",
)

iface.launch(share=True)
```
